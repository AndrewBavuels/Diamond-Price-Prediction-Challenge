```python
diamonds_train.drop(columns=['index_id','city_id','city_id.1','index_id.1','index_id.2',
                      'cut_id','color_id', 'clarity_id', 'cut_id.1', 'color_id.1','clarity_id.1'], inplace=True)
```

```python
column_order = ['city', 'cut','color', 'clarity','carat','depth','table','x', 'y', 'z','price']
diamonds_train = diamonds_train[column_order]
diamonds_train.head()
```

```python
diamonds_train.dtypes
```

```python
diamonds_train.columns
cat_diamonds_train = ['city', 'cut', 'color', 'clarity']
```

```python
cat_diamonds_train = diamonds_train[['city', 'cut', 'color', 'clarity']]
lab_enc = LabelEncoder()
cat_lab_enc = {}

for column in cat_diamonds_train.columns:
    cat_lab_enc[column] = lab_enc.fit_transform(diamonds_train[column])

cat_lab_enc_df = pd.DataFrame(cat_lab_enc)
cat_lab_enc_df
```

```python
diamonds_train = pd.concat([cat_lab_enc_df, diamonds_train.drop(columns=['city', 'cut', 'color', 'clarity'])], axis=1)
diamonds_train
```

```python
from pathlib import Path
filepath = Path('diamonds_train.csv', index=False)  
filepath.parent.mkdir(parents=True, exist_ok=True) 
diamonds_train.to_csv(filepath, index=False, encoding='utf-8')
```
