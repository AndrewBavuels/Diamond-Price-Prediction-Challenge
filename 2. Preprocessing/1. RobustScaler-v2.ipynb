```python
import matplotlib.pyplot as plt
import random
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
diamonds_train = pd.read_csv('diamonds_train_carat_clean.csv')
diamonds_train.head()
```

```python
rob_scaler = RobustScaler()
diamonds_train_scaled = diamonds_train.copy()  # Creamos una copia del dataframe original
diamonds_train_scaled.iloc[:, :-1] = rob_scaler.fit_transform(diamonds_train.iloc[:, :-1])
# Mostramos los primeros datos de los datos de prueba escalados
print("\nDiamonds Train Scaled:")
diamonds_train_scaled.head()
```

```python
diamonds_train = diamonds_train_scaled.copy()
```

```python
from pathlib import Path
filepath = Path('../16.02. exports/diamonds_train_rob_scl_clean.csv', index=False)  
filepath.parent.mkdir(parents=True, exist_ok=True) 
diamonds_train.to_csv(filepath, index=False, encoding='utf-8')
```

```python
diamonds_test = pd.read_csv('diamonds_test_carat_clean.csv')
diamonds_test.head()
```

```python
# Escalamos los datos de prueba usando el scaler ya ajustado

diamonds_test_scaled = diamonds_test.copy()
diamonds_test_scaled.iloc[:, :] = rob_scaler.transform(diamonds_test)

# Mostramos los primeros datos de los datos de prueba escalados
print("\nDiamonds Test Scaled:")
diamonds_test_scaled.head()
```

```python
diamonds_test = diamonds_test_scaled.copy()
```

```python
from pathlib import Path
filepath = Path('../16.02. exports/diamonds_test_rob_scl_clean.csv', index=False)  
filepath.parent.mkdir(parents=True, exist_ok=True) 
diamonds_test.to_csv(filepath, index=False, encoding='utf-8')
```
